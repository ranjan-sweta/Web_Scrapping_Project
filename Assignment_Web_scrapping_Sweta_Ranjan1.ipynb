```python
#Importing Libraries

# Pandas to put in Data Frame  and then in excel.
# Request to request the fetch request and send to html server to retrieve the Data.
# Beautifulsoup to fetch the data from the website.

import pandas as pd
import requests
from bs4 import BeautifulSoup 
```

```python
Content = [] #Taking an empty list to store the output.

for i in range(1,6):
  url = f'http://quotes.toscrape.com/page/{i}/' #fetching the website with page numbers.
  response = requests.get(url) #connecting our python to the website.
  response = response.content #Getting the content of the website.
  soup = BeautifulSoup(response,'html.parser') #connecting to the HTML server.
  webpage = soup.find('div') #getting the class div.
  contents = webpage.find_all('div', class_ = 'quote') #here, getting all the content under div where class is 'quote'.
  for content in contents:
    author = content.find('small',class_ ='author').text
    quotes = content.find('span').text
    tag = content.find('meta',class_='keywords')
    quote_tag = tag.attrs['content']
    Content.append([quotes,author,quote_tag])
```

```python
df = pd.DataFrame(Content,columns = ['Author', 'Qoutes','Tags' ]) #Making DataFrame
```
